The idea : Make a simulation with cell size that are equal to sample plots in area. Find all sample plots from MAGPlot (or from other sample plot database from Yan Boulanger) in the study landscape that have been undisturbed for years/decades and for which we have data. Translate the individual tree biomass in the sample plot database into cohort age + biomass structure.
Then, for each sample plot we have, initialize a cell in LANDIS-II with the beginning cohort structure. Let the model run so that every sample plot growth is simulated as pixels for the years we have for each sample plot. Compute the difference in trajectory of biomass throughout the simulation. Identify sample plots for which the dynamic of the simulation didn't match (more than 20%), and plot them to identify what went wrong.
