## Empirical data for biomass growth curves.

This is an essential part of the calibration process, especially the first step in which we will calibrate each species in "ideal" conditions of growth (mild climate, good soil water retention, etc.). These "ideal" conditions are important to isolate the effect of other parameters and other sources of mortality, so that we can focus on the basic parameters of growth and senescence for your species (see [the first calibration step](./6.Monoculture_ideal_conditions.ipynb)).

Basically, what you want is data that allows you to infer a biomass growth curve for each given species, showing the evolution of the biomass of a cohort for the species in conditions that are as ideal as possible.

As of today, I've ran accross three different strategies to get this data. I'll describe them in the following sections.

### Option 1 : Sample plot data.

By looking at individual tree data from sample plot data, you can infer tree biomass and tree age, making each tree measurement or plot measurement a point on your graph.

By looking at the points that are at the top of the cloud of points that will result, you can get and idea of the maximum biomass a tree can reach for a given age, which should give you good information about the growth curve. This is a method used by many research so far.

The problem is that sample plot data is often messy, full of outliers, biases, needing sharing agrements, etc. As such, it will require important work to clean, and several important assumptions will influence your results enormously. For example, the "tree age" data is often missing, and you'll have to infer it from DBH; and you'll also have to choose how to translate your individual tree biomass into biomass/surface (as all LANDIS-II biomass output is interpreted in biomass/surface, not biomass/tree).

This data can still be quite useful to derive certain caractestics of the growth curve (e.g. the age of the peak of growth for a given species). 

In Canada, aquiring sample plot data has been very much facilitated by the publication of the [MAGPlot database](https://open.canada.ca/data/en/dataset/8824392d-464e-413d-8bde-eaed61c79743). It contains standardized and unified plot data accross all canadian provinces, with tree DBH as the best measure available at the plot level. Some of the data (e.g. plot coordinates) are randomized to preserve sensible information.

### Option 2 : Tree growth models.

These models are complex equations develloped by foresters, industries or governmental agencies to try and predict growth or yield for different species in different conditions. These are often calibrated based on sample plot data; as such, the work of cleaning/dealing with sample plot data is often done for you. An example is the famous [Forest Vegetation Simulator](www.fs.usda.gov/fvs/) from the USDA. Other such models are simply growth curves published in the litterature, sometimes in books.

The curves produced by these models are often in biomass/surface, which makes things easier to compare with LANDIS-II outputs than sample plot. You can also tweak the parameters of these models so that they predict biomass growth in conditions as ideal as possible, making it close to what you're looking for.

The main issue is that these models can still produce pretty unrealistic curves; might not be available for your study area/region; and might still need parameters derived from sample plot data (e.g. Site index and maximum basal area for FVS). So while they can make your life simpler, they can still be difficult to use.

FVS can be compiled into a Dockerfile and use in Python; I've tried it at first and will keep the functions that I've written in a separate document for you to re-use if necessary.

Be warned that the decreasing part of the curves produced by these often often don't include mortality by senescence, which can make the cruve growth forever, making it difficult to identify the biomass peak which is very important for calibration purposes.

Tree growth models have been used by several research teams, including Brian Sturtevant's and Eric Gustafson (the latter being the author of the calibration tips for PnET-Succession that are the basis of these documents).

### Option 3 : Forest inventory data.

Some forest inventories contain data that allow you to infer tree or stand age and the biomass of your species of interest. For example, the [National Forest Inventory of Canada](https://nfi.nfis.org/en/) proposes raster maps of several forest attributes Canada-wide, including pixel/stand forest age, and the biomass of each individual species that are considered in the forest inventory.

If you use each pixel from the maps of a single year as a point and plot a cloud of points on a biomass/time axis, you might get a good estimation of a growth curve, with a growth peak and a initial growth (the two most important parts in our case). This is, in essence, a ["space-for-time" substitution](https://link.springer.com/chapter/10.1007/978-1-4615-7358-6_5), where you look at many different points in space that will be forests at different stages of growth, rather than looking at the long-term history of a few forests.

The big limitation of this data is that you often get one "stand age" or "pixel age" value per pixel, rather than a separate age value for the trees of a given species in the pixel. As such, in a pixel of 80 years, you cannot know if the biomass of species X comes from trees that are 80 years old, or if these trees are younger. But you can interpret the upper points of the cloud of points as "the maximum biomass that a species can reach for this stand age", which is pretty close to the mark of what you're looking for.

Another big limitation is that in most pixels (if not all), your species will be mixed with other species. This is especially true if you are calibrating a "companion" species (the type of tree species that is never abundant, but yet found in many forests). You can restrict the points of your data to only the pixels where more than 50% of the biomass comes from your species of interest; but you might not get enough points for some rarer species. In these case, you will probably need to develop a correction of biomass to try and have an estimate of the biomass that your species can reach when it composes 100% of the stand. You might do this by making a statistical model that attempts to predict the biomass of your species in a stand or pixel based on the age of the pixel, and the % of composition of the species. You might want to train your statistic model on only the top percentile of points (as you're looking out for the most biomass your species can achieve for a given age). 

This data is often generated through a combination of forest inventory sampling (aerial photography, etc.) and sample plot data (which are assignated to stand or pixels via assignation algorithms like the kNN method). As such, it often contains information from sample plots, but which has already been cleaned and dealt with.

The biomass data is often expressed in biomass/surface, making it easy to compare to LANDIS-II outputs.

This is the data I will be using here. From what I heard, other teams have been using such data in the past.

### Producing estimated growth curves in ideal conditions (in Canada) for our species with NFI data using a Generalized Additive Model

#### Why we are using a Generalized Additive Model to produce estimates of growth curves in ideal conditions

Based on the NFI data shown and explored in the previous section, we now have to produce estimated growth curves based on this data that we will be able to use as a "target" for the first step of the calibration of PnET-Succession, where we will try to match the curves produced by PnET-Succession with the data.

Since the first calibration step is based on growth curves in "ideal" conditions, where the species we are calibrating have no competition for water and for light, we especially have to estimate the growth curves for monospecific stands, where the relative abundance of the species is 100%. Unfortunatly, pixels that contains only one species are very rare in the NFI rasters since the pixels are 250x250m, and also because in nature, monospecific stands can be quite rare for many species. Even worse, some species have a very low amount of pixels in which they have a high relative abundance, or even none at all. A good example here is *Acer rubrum*, the red maple, which has a low amount of points with relative abondance above 50% when compared to the two other species we are calibrating here, *Abies balsamea* and *Picea mariana*. I expect that many other tree species available in the NFI will have the same issue. 

As such, I propose to use statistical modelling to use the information contained in the pixels we do have (where the relative abundance of any given species is below 100%) in order to generate growth curves for different level of abundances, and especially for a 100% abundance. I set up to use Generalized Additive Models, as they model non-linear effects of variables quite well, and because I expect the effect of the factors that will be used in this model (forest age and relative species abundance) will be non-linear.

In this approach, we will treat every pixel from the NFI raster as a single data point associated with 3 variables, one for the different raster : age of the pixel; biomass of the species X in the pixel; and relative abundance (in %) of species X in the pixel. By grouping these points into age windows, we can then use a GAM to predict the upper percentile of biomass of species X observed for a given age window, and for a given % relative abundance. As such, by predicting the upper percentile per age window, the GAM will predict how much maximal biomass species X can achieve for a given age, and a given % relative abundance.

Once the GAM is fitted, we make predictions for a 100% relative abundance, which is what we want. These predictions will not be perfect, as very few (if none) pixels with a 100% relative abundance exist, as said before; and as such, the model will be extrapolating outside of the data on which it has been fitted. But whatever the approach we use (i.e. using the FVS model, or using sample plot data), the problem will be the same : as monospecific stands are rarely observed in nature, any estimate that we're trying to get here will be based on an extrapolation one way or another. This is the limitation of using a first calibration step in "ideal conditions", although using this approach has some great advantages. For more about this, take a look at [0.1.Some_important_notions.ipynb](./0.1.Some_important_notions.ipynb) where this question is talked about in details.

Having these GAMs fitted will thus allow us to produce growth curves predicted by the GAM for different level of % relative abundance for the species X, including the 100% curve that we are looking for.

#### Methodology : using GAMs on pre-treated data from the NFI, and selecting the most realistic curve produced for the most limited spatial extent of data

I've made some first attempts fitting some relatively simple GAMs on the data of the NFI for the three species I'm calibrating here. Unfortunatly, this didn't go well : the growth curves produced by the GAM was quite non-sensical, and didn't fit the points very well. I'm not an expert in statistics, but I expect that this was because the model being fitted on patterns within the data that I couldn't see with my human eyes (as in my eyes, the growth pattern visible in the data was quite clear). Maybe it's the overabundance of low-abundance pixels, and also the presence of some high-abundance pixels with relatively low biomass values and other outliers tended to create unexpected and absurd GAM behaviour.

I solved this issue by pre-processing the data (see figure below) to 1) reduce the number of points, 2) remove the outliers, and 3) force a monotonic increasing rule regarding the effect of abundance of the high percentiles of biomass. This pre-processing is done by defining windows of age and % relative abundance; in each of these windows, we gather all of the pixels/points that are inside of it (i.e. pixels with an age and a % relative abundance for species X that have the value for the window); and then, we compute the upper percentile of biomass for the pixels/points in this window. We repeat the process accross all age/% relative abondance windows possible (without superposing the windows).

Furthermore, the function can remove points that break the monotonic increasing tendancy of the effect of the abundance on the biomass; i.e. if a percentile for an abundance window of 80 is lower than the percentile of the previous abundance window (e.g. 75) for the same age window, we will not take it into account. This is done to clarify the tendancy of the effect of the abundance on the biomass to help the GAMs fit better on the data, using the assumption that points that do not follow this tendancy are deemed to be errors or extremes.

Here is an depiction of the procedure in visual form :

![](./Images/GAM_DataProessing_NFI_Method.PNG)

This pre-processing is highly customizible, allowing you to define the size of the windows, the percentile computed, or to exclude certain pixels; see the function `createPercentilesBiomassDatasetsFromNFIData` in `functionsForCalibration.py` . You can put the argument `printPlot` as `True` in order to see the resulting points datasets.

The function `createPercentilesBiomassDatasetsFromNFIData` will generate four points datasets that correspond to four different spatial extent where we take the data from the NFI : the study landscape we are calibrating for; the ecozones that directly intersect with the study landscapes; the same ecozones + the ones that surrounds them in Canada; and finally, all of the data from the NFI for all of Canada. This is done so that we can fit GAMs and produce estimated growth curves for these different spatial extents, in order to compare the curves we obtain. The goal will be to select a curve that seems as realistic as possible while trying to select data as close to the study landscape as possible, in order to avoid mixing a lot of intra-specific variability (different phenotypes, variants, etc.) from the species in our dataset. However, sometimes, the data from the study landscape will not be enough to produce realistic curves, because there is not enough data points for the species in it (especially points with enough % relative abundance for the species).

As such, the function `getGamDiagnostics` (used after `createPercentilesBiomassDatasetsFromNFIData`) will compute some diagnostic values for the 100% abundance curves produced by the GAMs fitted on each of these datasets, and then use a simple algorithm to determine the best curve, and export it to a `.csv` file where it can be used in the subsequent notebooks for the calibration. The selection is made by first determining which curves are considered "valid" according to three diagnostics : a pseudo R2 > 0.85; a curve shape that is truly concave; and a percentage of negative biomass values along the curve (for the entire age range of the curve) being inferior to 20%. Then, we simply select the first valid curve in the subset of points that is the smallest (i.e. closest to the study landscape), in order to reduce the intra-specific variability present in the subset of points on which our estimation is based.

#### The GAM equation we will use here

The most robust GAM structure I've found after many trials accross the data for the three species we're calibrating here is the following. It's a GAM structure that produces mostly relatistic curves (e.g. curves that are concave, with clearly defined growth peaks, and start near 0) accross all species and data subsets (e.g. study landscape, ecozone, surrounding ecozones, and Canada).

$\text{High Biomass Percentile} = s(age, by(\% abundance), \text{with a concave constraint}) +  s(\% abundance, by(age), \text{with a concave constraint}) + s(age, \text{with a concave constraint}) + s(\% abundance, \text{with a concave constraint})$

Here, the `by` statement indicates a simple interaction between the two factors (see [the documentation of the pyGAM package](https://pygam.readthedocs.io/en/latest/notebooks/tour_of_pygam.html)). We can also use a more complex interaction using a tensor (`te()`), but all of my attempts using tensors resulted in curves that didn't make sense. It seems that the tensor was too sensitive to missing data, and tended to dip quickly into a negative effect, leading to non-sensical curves.

The curves have been created using different sources of remote sensing and aerial laser data. Furthermore, these curves are develloped inside 150x150km tiles accross Canada, with one curve per tile (as long as enough data for a given species was available in the tile).

We can use the data from the article, which is freely available in a very useful R package (see [ptompalski/RSYC](https://github.com/ptompalski/RSYC)), in order to assess the estimates we have obtained with the GAMs.

:::{warning}
Keep in mind that the biomass curves produced by Tompalski et al. 2025 are made with a different methodology than what we're doing here. In particular, their curves represent the biomass of a stand **dominated** by a given species; but they don't seem to remove the biomass from other species that will be inside the stand with the dominating species. Here, thanks to the NFI relative abundance raster, we do isolate the biomass for a given species. As such, species that can be dominant but which are still often not "very dominant" (not a lot of % relative biomass) will have their biomass mixed in with other species in the curves of Tompalski et al. 2025. This makes it that the curves of Tompalski et al. 2025 will often have higher biomass peaks than what is computed here for a given species; especially for species which display a low relative abondance accross most of their pixel because they are often companion species, like *Acer rubrum*.

Also, the curves of Tompalski et al. 2025 often display no scenescence/decrease in biomass, because they focus on the stand level and not on the species in the stand. Meanwhile, or method based on the NFI data does track the scenescence of given species inside a stand more easily, since we can isolate the biomass from the species of interest. 
:::

#### Limitations of the approach with GAMs to estimate growth in ideal conditions

The approach used here has a lot of advantages (replicable, using open-source and peer-reviewed datasets that are easily available, can be adapted to future datasets of the same format, etc.). But it also have important limitations. In particular, two things have to be kept in mind :

- We are using a statistical model to exprapolate, which means that the estimates that we get here are nothing but estimates.
- Data from the NFI is on 250x250m cells, and is generated through k-NN association with sample plot data; what this means is that there is a lot of averaging going on. As a consequence, if you compare the estimate of biomass in tons/ha obtained here with sample plot data, you will tend to see much higher maximum biomass measure from the sample plot data, since there a lot more extreme values than in the NFI data.

The most important thing in the end is to make sure that the estimates of growth in ideal conditions we got here represent the differences between species quite well : whatever the values themselves, the ranking of species in terms of their growth capacity should be similar to their biological reality. It's hard to consider that when we mix it with the fact that there is an important intra-specific variability that can mix things up; but overall, if you see one species that is supposed to be not very competitive achieving a lot of biomass compared to another one that is know to be very competitive or to achieve a lot of biomass, you might want to take a look at your curves again, and maybe select a different spatial extent or dataset.

Still, I consider this approach to be quite solid, especially after I contacted several researchers to get their opinion on it.

## Values needed to estimate certain parameters

### Longevity and Max Longevity

What we will call here "Longevity" and "Max Longevity" are necessary to calibrate the senescence part of the growth curve for our species.

The difference between the two is this, per the team of Brian Sturtevant who explained their methodology to me :

- Longevity is the rule-of-thumb, expert opinion longevity for a given tree species. This is the age at which it's expected that a tree of a given species will die, in a give region. These are the number often reported in the litterature, and also used in the core species parameter file of LANDIS-II as its "Longevity" parameter for each tree species used.
- Max Longevity is the real, empirical maximum age that a tree species can reach in the study area. It's often not reported and difficult to find. A good way to find it is to look at sample plot data and find the oldest trees ever found for a given species (which can be difficult to find as the "age" data is often not directly available in sample plots data).

Brian Sturtevant and his team have used these two notions to calibrate the senescence part of the growth curves in ideal conditions as explained in [./6.Monoculture_ideal_conditions.py](./6.Monoculture_ideal_conditions.py).  This part from [this supplementary material of Reese et al. (2025)](https://cdnsciencepub.com/doi/suppl/10.1139/cjfr-2024-0085/suppl_file/cjfr-2024-0085suppla.docx) describes Brian's methodology of explains it all :

> We used an unpublished review and analysis led by John Du Plissis at the University of Minnesota-Duluth to define both tree species longevity (the age at which a species age cohort can no longer persist) and the initiation of decline that is defined by the age-related photosynthesis reduction factor (PsnAgeRed) parameter in PnET Succession. In general, the maximum species longevity of Loehle (1988) corresponds with the tree species longevity parameter, while the typical species’ longevity from the same publication determines the initiation of decline, where PsnAgeRed is calibrated such that the fAGE (age reduction factor) begins to decline (~0.99) at the Typical Mortality age of the species.


The data used by Brian's team for Typical Longevity and Max Longevity are cited as coming from https://sci-hub.st/https://cdnsciencepub.com/doi/10.1139/x88-032 in ReeseEtAl_CanJForRes_2024_supplA.docx ; it's an article that seem to compute them based on several characteristics of tree species.

In the end, Brian's team is using a combination of data from [Loehle (1988)](https://cdnsciencepub.com/doi/10.1139/x88-032) and estimates found by John Du Plissis (yet unpublished). I'm going to use the data from [Loehle (1988)](https://cdnsciencepub.com/doi/10.1139/x88-032), which should be enough for our goals until the data from Du Plissis is published.

[Loehle (1988)](https://cdnsciencepub.com/doi/10.1139/x88-032), defines "Typical" versus "Maximum" longevity in the following terms :

> Typical longevity is the age of the older trees in an old-growth stand, whereas maximum longevity is the upper known maximum above which very few trees are found.

We can find the value of typical longevity and maximum longevity from the Appendix 1 of [Loehle (1988)](https://cdnsciencepub.com/doi/10.1139/x88-032) (see end of the .pdf). Here are the values for our three species :

### Temperature tolerance/species optimal temperature

To create expected outcomes of competitions between species in conditions of increasing temperature, we need to have some idea of the either the tolerance of each tree species to higher temperature, or of the temperature they prefer to grow well.

Unfortunately, I wasn't able to find any dataset regarding the temperature tolerance of best growth temperatures of tree species in Canada. While the TRY database have traits linked to temperature tolerance or preference, there seem to be relatively few measurements for Canadian tree species.

I've decided to go with an alternative route, and to compute estimates of the favorite growth temperature of species based on a simple assumption : where the tree species is most present/abundant in Canada, then the temperature is best for it. This is an imperfect assumption of course, as soils, disturbances and competition with other species also affect its presence; but I'll consider it enough to produce reliable estimates that will give us a broad idea of the temperatures the species prefer.

Temperatures and precipitations are, according to the litterature, two of the most important factors explaining the distribution of tree species on earth. However, the effect of temperature on trees can appear in different ways, sometimes indirect; for example, by killing out insects or diseases that will attack the tree in other regions. As such, it's not because a tree species is most abundant in a place with a given temperature that this means that this temperature is optimal for its growth : it might be that other effects of the temperatures (and also of precipitations) are at play to help the species grow in this place.

Here, we want to get a temperature for each tree species that reflects their relative "sweet spot" of growth (where their temperature is optimal for growth), which will help us predict the outcome of competition between our tree species in condtions where temperatures (and shade) are the only factors of competition. Ideally, we would get these through expeximentation, where we can put tree saplings in controlled conditions and with different temperatures in order to see how they respond to it. This would allow us to isolate the direct effect of temperature on tree growth, independant of other factors (i.e. disturbances, solar radiation, precipitations, etc.).

Unfortunatly, we do not have such experimentation data for tree species in Canada. But we do know that the direct effect of temperature on tree growth is one of the forces shapping the repartition of tree species in Canada. As such, I make the assumption that by looking at the average summer temperature accross Canada that we then average by ponderating it by the distribution of the biomass of the species accross Canada, we will get a value that is pretty close the optimal growth temperature of the species.

:::{tip}
To read more about the limitations of this approach, and especially the lack of correlation between the localisation of a species and its optimal temperature, see [here](https://academic.oup.com/treephys/article-abstract/30/6/669/1614202).

Also, you can read about the fact that while temperature as an effect on tree growth, in nature, this interact very closely with water availability, and that the notion of "optimal temperature" thus tends to change with the availability of water; see [here](https://pubmed.ncbi.nlm.nih.gov/31883292/).

So again, this approach has a lot of limitation; but I will still use it to try and catch any strange behavior from the model when species are competiting with each other in changing temperature conditions.
:::

```python
print("\nStep 1: Loading Ouranos climate data with latitude subset...")
print("  Opening OPeNDAP dataset...")

ds = xr.open_dataset(climateDataURL)

# Get coordinate names
# Allows us to adapt to more formats of 
lon_name = 'lon' if 'lon' in ds.coords else 'longitude'
lat_name = 'lat' if 'lat' in ds.coords else 'latitude'

# Subset to latitudes >= 40°N【0】
# I do this because subsetting with a polygon that corresponds to Canada
# or to the extent of the NFI raster has proven to be quite the headache.
# This is simpler and it works very well.
print("  Subsetting to latitudes >= 40°N to reduce the data downloaded")
if ds[lat_name].ndim == 1:
    # 1D coordinates - simple selection
    ds_subset = ds.where(ds[lat_name] >= 40, drop=True)
else:
    # 2D coordinates - need to check any latitude value in row
    lat_mask = (ds[lat_name] >= 40).any(dim=ds[lat_name].dims[1])
    ds_subset = ds.isel({ds[lat_name].dims[0]: lat_mask})

print(f"  Subset dataset shape: {dict(ds_subset.dims)}")

# Filter for 1962-2011 and summer months
print("  Filtering temporal range (1962-2011) and summer months...")
ds_filtered = ds_subset.sel(time=slice('1962-01-01', '2011-12-31'))
summer_months = ds_filtered['time.month'].isin([6, 7, 8])
tas_summer = ds_filtered['tas'].isel(time=summer_months)

# Calculate mean summer temperature (convert K to °C)
print("  Computing mean summer temperature...")
mean_summer_temp = tas_summer.mean(dim='time') - 273.15

# Load into memory
print("  Loading climate data into memory...")
mean_summer_temp = mean_summer_temp.load()

print(f"  Climate data shape: {mean_summer_temp.shape}")
print(f"  Temperature range: {float(mean_summer_temp.min()):.2f}°C - {float(mean_summer_temp.max()):.2f}°C")
```

### Computing expected competition outcomes between species

After the first calibration step (where we simulate an ideal situation for each tree species with no competition, good soil and enough precipitations for water, etc.), we will do several more step where we will check if the model produces "realistic" competition outcomes of species against species in several conditions :

- Competition for shade (two species together with a good water input through soil and precipitation)
- Drought (two species together but with a lack of water)
- Waterlogging (two species together but too much water)
- Temperature (two species together with increased temperatures)

This is inspired by this text from the [calibration tips of Eric Gustafson](./ReferencesAndData/Documentation/Gustafson2024PnETUserGuide.pdf) :

> It is prudent to test the competitive behavior of pairs of species to ensure that the behavior is appropriate. I recommend that you conduct some tests of two and/or three species combinations using your calibration setup to ensure that species compete as you would expect. Extremely poor competitive ability usually results only when a cohort is inappropriately overtopped by another and becomes lightstarved. (...) It is highly recommended that you verify your calibrations by simulating several similar species together to ensure that they in fact compete as expected. If not, often a slight tweak of one parameter can bring them in line with expectations.

Ideally, we will test every species against another, and see which one "wins" by having the most biomass in the pixel after a long time (100 years), and compare this against "expected outcomes". Expected outcomes are predictions we will make in advance based on what we know of the biology of each tree species. For example, if we put a shade tolerant species with a shade intolerant species in competition in otherwise ideal conditions of water and temperature, we expect the shade tolerant species to "win" on the long term. If the model cannot produce the expected outcomes, that might mean that we haven't represented the differences between species in the right way (e.g., their relative difference in shade tolerance is not implemented well enough in our parameters). But getting these inter-species different is crucial according to [Eric Gustafson](./ReferencesAndData/Documentation/Gustafson2024PnETUserGuide.pdf) so that the model can properly represent the competition outcomes between them.

But it's not that simple. Here, the challenge comes in defining the expected outcomes (how do we know which species is going to win ?) and what "winning" means.

- Defining a win is difficult because of the variability that comes with it. E.g., a shade tolerant species versus a shade intolerant one might reach more biomass than the other after 100 years; but maybe not if the cohorts starts at very different age, or with very different biomass. For example, the competiton could start with a very large and old shade intolerant cohort, which will take the shade tolerant species a long time to overcome.
- Defining the expected outcomes (who wins) is also difficult because we have no solid empirical data on the competition between species; or at least, not enough empirical data to quantity how easy or how frequent it will be to win for a species against another. How do we know if *Acer rubrum* will tend to "win" against *Abies balsamea* in a competition for light one 100 years of simulation ? It's doubly difficult because empirical data comes from a place where many competitions and factor affect tree growth at once (shade, water, temperature, etc.), along with disturbances (which we're not including here).

As such, the best we can do here is to use arbitrarely-defined expected outcomes (although based on the real and empirical/know difference in shade tolerance, drought tolerance, etc...of our species), and use arbitrarely-defined definitions of a "win" and of a threshold above which we consider that there is too much difference between the expected outcome of the actual ouctomes of the competition from the simulations.

To define the expected outcomes, I propose to use the tolerance traits that can be found in the [TRY database](https://www.try-db.org/TryWeb/Home.php) : Shade tolerance, drought tolerance, waterlogging tolerance, as well as the optimal temperature we calculated in a previous section. These traits are often used in other LANDIS-II studies (see supp. material from [Reese et al. 2024](https://cdnsciencepub.com/doi/10.1139/cjfr-2024-0085)) to calibrate PnET-Succession of produce expected outcomes.

Here is the arbitrary methodology that I propose to use :

- We define a "win" when one species has achieved more biomass than the other after 100 years of simulation.
- To define the intensity or the variation in "winning" for one species competing against another, we will make many simulations where we start with cohorts of different ages for the two species. We will compute the amount of cells in which, after 100 years of simulation, the "challenger" species has more biomass than the "opponent" species. This amount will correspond to the intensity of "wins". For example, if 60% of cells have more biomass for the challenger species after 100 years of simulation, then we will say that it has a win rate of 60%.
- To define the expected outcomes, we will do a simple computation based on the tolerance traits, but always account for the competition for shade (which is still there despite the inclusion of drought, waterlogging or changing temperatures)
    - For shade, the expected win will be derived linearly from the following rules :
        - If the challenger species has a tolerance of 5 (very tolerant) and the opponent has a tolerance of 1 (very intolerant), then we expect a 100% winning frequency for the challenger.
        - If the oppososite happens (1 for the challenger, 5 for the opponent), we expect a 0% winning frequency.
        - If the challenger and the oponnent have the same shade tolerance, we expect a 50% winning frequency.
        - For any other combination of shade tolerance between challenger and opponent, we derive the winning frequency linearly from the previous rules (`prob_win = 0.5 + 0.125 * diff`)
    - For drought and waterlogging, the rules are as follow :
        - Simulations for drought will be done with an excessively drained soil that doesn't retain water much (e.g. coarse sand), as per the supp. material A from [Reese et al. 2024](https://cdnsciencepub.com/doi/10.1139/cjfr-2024-0085)
        - Simulatons for waterlogging will be done with a soil that retains water a lot (e.g.. mucky peat), as per supp. material A from [Reese et al. 2024](https://cdnsciencepub.com/doi/10.1139/cjfr-2024-0085)
        - We first compute the frequency of winning with the same rule à with the shade tolerance (see above), but with the drought tolerance or waterlogging tolerance used in the calculation instead.
        - Then, we also compute the frequency of winning the shade competition (based on the shade tolerance) with the same rule as before.
        - We then compute the final expected frequency of winning the competition by doing a ponderated average of the expected frequency of winning the shade competition, and the expected frequency of winning the drought or waterlogging competition. In that way, we still account for the shade competition happenning.
    - For temperatures, the rules are as follow :
        - If the challenger species has an optimal temperature (as calculated in a previous section) that is 5 degrees celcius or more closer to the average summer temperature that will be use in the competition simulation when compared to the opponent species, then it has 100% change winning. For example : if the average summer temperature in the simulation is 12 degrees celcius, and the optimal temperature of the challenge is 11 degrees while the one of the opponent is 6 degrees celcius, then the challenger species has an optimal temperature 5 degrees closers to the average summer temperature in the simulation, and is therefore predicted to have 100% change winning.
        - If the opposite happens, then we expect a 0% winning frequency for the challenger.
        - We then compute the final expected frequency of winning the competition by doing a ponderated average of the expected frequency of winning the temperature competition, and the expected frequency of winning the drought or waterlogging competition. In that way, we still account for the shade competition happenning.
